# Scrap articles from web without any pre-processing

```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
input_file_name = 'train_spam_url.csv'
output_file_name = 'train_spam_raw.csv'
```

```python
data = pd.read_csv(input_file_name)
```

```python
data.head()
```

```python
data.shape
```

```python
# Remove Duplicates if any from url file.
data.drop_duplicates(subset='URL', inplace=True )
```

```python
data.shape
```

```python
# Create an empty dataframe to store scrapped values.
data_scrapped = pd.DataFrame(columns=['URL', 'Articles'])
```

```python
# For each URL, visit web and scrap data. Save the scrapped data to a dataframe
good_url = 0
bad_url = []
count = 0

for url in data['URL']:
    try:
        html = urlopen(url).read().decode('utf8') # Convert url to html
        soup = BeautifulSoup(html, 'html.parser') # Create Soup
        paragraphs = soup.select('p')
        if len(paragraphs) > 0:
            data_scrapped = data_scrapped.append({'URL': url, 'Articles': paragraphs}, ignore_index=True)
            good_url+=1
        else:
            bad_url.append(url)
    except:
        bad_url.append(url)
    
    count+=1
    print('Completed: {}'.format(count))
```

```python
print('good url: {}'.format(good_url))
```

```python
print('bad url: {}'.format(len(bad_url)))
```

```python
# Save dataframe for further processing.
data_scrapped.to_csv(output_file_name, index=False)
```
